```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = Image.open("assets/tileset.png")
image = np.array(image)
col = 13
row = 13
for r in range(row):
    ridx = r * 17
    for c in range(col):
        cidx = c * 17
        print(f"{cidx} / TEX_X_PIX_CNT, {ridx} / TEX_Y_PIX_CNT, {cidx + 16} / TEX_X_PIX_CNT, {ridx+16} / TEX_Y_PIX_CNT")
        plt.figure()
        plt.title(f"{cidx}, {ridx}, {cidx + 16}, {ridx+16}")
        plt.imshow(image[ridx:ridx + 16, cidx:cidx+16, :])
        plt.show()
```

```python
rstart = 15 * 17 - 1
# rend = 16 * 17
for c in range(col):
    cidx = c * 17
    print(f"{cidx} / TEX_X_PIX_CNT, {rstart} / TEX_Y_PIX_CNT, {cidx + 16} / TEX_X_PIX_CNT, {rstart+16} / TEX_Y_PIX_CNT")
    plt.figure()
    plt.title(f"{cidx}, {rstart}, {cidx + 16}, {rstart+16}")
    plt.imshow(image[rstart:rstart + 16, cidx:cidx+16, :])
    plt.show()
```
